```python
#categorise weight category coulmn
df2 = df1
```

```python
for ind in df.index:
    if df2["bmi"][ind] < 18.5:
        df2["weight category"][ind] = "Under weight"
    if df2["bmi"][ind]>=18.5 and df2["bmi"][ind]<24.9:
        df2["weight category"][ind] = "normal weight"
    if df2["bmi"][ind]>=24.9 and df2["bmi"][ind]<30.0:
        df2["weight category"][ind] = "Over weight"
    if df2["bmi"][ind]>=30.0:
        df2["weight category"][ind] = "obese"
```
